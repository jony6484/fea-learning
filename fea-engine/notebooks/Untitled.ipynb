```python
df = pd.read_csv('../data/fea_tests_2_5_bar_15_08_2021.csv')
```

```python
df
```

```python
cols = df.columns
cols_C = cols[cols.str.startswith('C')]
cols_heights = cols[cols.str.contains('height')]
cols_pressure = cols[cols.str.contains('pressure')]
```

```python
def fun(cols):
    x = cols[cols_heights]
    y = cols[cols_pressure]
    return np.polyfit(x, y, deg=4)
```

```python
df.apply(fun, axis=1)
```

```python
df_coef = pd.DataFrame(zip(*df.apply(fun, axis=1))).T
```
